```python
# Pandas
import pandas as pd
import numpy as np
```

```python
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
#handling missing data 
cleaning_df = df.dropna()
cleaning_df
```

```python
# Corrected code with fixed column name spelling and proper parentheses
df['Energy Consumption (MWh)'].fillna(df['Energy Consumption (MWh)'].mean(), inplace=True)
df['Cost (Million $)'].fillna(df['Cost (Million $)'].mean(), inplace=True)
df
```

```python
#cost per MWh
df['Cost per MWh'] = df['Cost (Million $)'] / df['Energy Consumption (MWh)']
df
```

```python
#cost per MWh
df['Cost per MWh'] = df['Energy Consumption (MWh)'] / df['Cost (Million $)']
df
```

```python
# data preprocessing
from sklearn.preprocessing import MinMaxScaler  
scaler = MinMaxScaler()
df[['Energy Consumption (MWh)', 'Cost (Million $)']] = scaler.fit_transform(df[['Energy Consumption (MWh)', 'Cost (Million $)']])
df
```

```python
from sklearn.preprocessing import StandardScaler  
std_scaler = StandardScaler()
df[['Energy Consumption (MWh)', 'Cost (Million $)']] = std_scaler.fit_transform(df[['Energy Consumption (MWh)', 'Cost (Million $)']])
df
```

```python
#OHE(one hot encoding)
df_energy_encoded = pd.get_dummies(df,columns = ['Energy Source'])
df_energy_encoded
```

```python
#visulisation
#line plot
import matplotlib.pyplot as plt
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
energy_cons = [1200, 1300, 1100, 1500, 1400, 1600]
plt.plot(months, energy_cons, marker = 'o', linestyle = '-', color = 'purple')
plt.xlabel('months')
plt.ylabel('energy cons')
plt.show()
```

```python
#barchart
energy_source = ['solar', 'wind', 'hydropower', 'biomass', 'geothermal', 'nuclear']
values = [1200, 400, 800, 1500, 700, 1600]
plt.bar(energy_source, values, color = 'purple', linestyle = '-', edgecolor = 'black')
plt.title('energy consumption by sources')
plt.xlabel('energy_source ')
plt.ylabel('values')
plt.show()
```

```python
plt.pie(values,labels=energy_source,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
#scatter plot
carbon_emission=[200,500,450,300,350,420]
plt.scatter(values,carbon_emission,color='purple')
plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emission (tons)')
plt.show()
```

```python
#custamized baeplot
plt.bar(energy_source,values,color='purple',edgecolor='black',linewidth=1.5)
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.title('Energy Consumption by Source')
plt.grid(True,linestyle='--',alpha=0.7)
plt.legend(['Energy Consumption'])
plt.show()
```
